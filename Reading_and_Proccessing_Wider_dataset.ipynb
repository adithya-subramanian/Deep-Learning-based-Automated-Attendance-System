```python
import numpy as np
import pandas as pd
import os
from multiprocessing import Pool
import multiprocessing

cpu_count = multiprocessing.cpu_count()
print(cpu_count)
```

```python
#print(os.getcwd())
loc = os.getcwd()+'/wider_face_split/wider_face_train_bbx_gt.txt'
File_object = open(loc,"r")
```

Attached the mappings between attribute names and label values.

blur: clear->0 normal blur->1 heavy blur->2

expression: typical expression->0 exaggerate expression->1

illumination: normal illumination->0 extreme illumination->1

occlusion: no occlusion->0 partial occlusion->1 heavy occlusion->2

pose: typical pose->0 atypical pose->1

invalid: false->0(valid image) true->1(invalid image)

The format of txt ground truth. File name Number of bounding box x1, y1, w, h, blur, expression, illumination, invalid, occlusion, pose

```python
name = []
Flag = 0
Counter = 0
Counter_list = []
img_data = []

with open(loc,'r') as file:
    
    for line in file:
        #print(line[-4:])
        if(line[-4:-1]=='jpg'):
            # print(line)
            name.append(line[:-1])
            Flag=1
        elif(Flag==1):
            Counter = int(line[:-1])
            Counter_list.append(Counter)
            #print(Counter)
            Flag=0
        else:
            #print(line[:-1].split())
            img_data.append(line[:-1].split())
```

```python
df = pd.DataFrame()
df['name'] = name
df['count'] = Counter_list
```

```python
df.head()
```

```python
name_1 = []
value_1 = []
cntr = 0

for i in range(0,len(df)):
    cntr +=1
    j = df.iloc[i,1]
    #print(j)
    while(j>0):
        name_1.append(df.iloc[i,0])
        value_1.append(img_data)
        cntr +=1
        j -=1
```

```python
img_df = pd.DataFrame(columns=['name','x1', 'y1', 'w', 'h', 'blur', 'expression', 'illumination', 'invalid', 'occlusion', 'pose'])

img_df['name'] = name_1
```

```python
img_df.head()
```

```python
x1 = []
y1 = []
w = []
h = []
blur = []
expression = []
illumination = []
invalid = []
oclusion = []
pose = []
```

```python
for lit in range(len(name_1)):
    x1.append(img_data[lit][0])
    y1.append(img_data[lit][1])
    w.append(img_data[lit][2])
    h.append(img_data[lit][3])
    blur.append(img_data[lit][4])
    expression.append(img_data[lit][5])
    illumination.append(img_data[lit][6])
    invalid.append(img_data[lit][7])
    oclusion.append(img_data[lit][8])
    pose.append(img_data[lit][9])
```

```python
img_df['x1'] = x1
img_df['y1'] = y1
img_df['w'] = w
img_df['h'] = h
img_df['blur'] = blur
img_df['expression'] = expression
img_df['illumination'] = illumination
img_df['invalid'] = invalid
img_df['oclusion'] = oclusion
img_df['pose'] = pose
```
