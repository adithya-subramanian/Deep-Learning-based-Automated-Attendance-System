```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Wider_stats.csv')
```

```python
df.head()
```

```python
df_summ_mean = df.groupby(['name']).mean().reset_index()
df_summ_mean.to_csv('df_summ_mean.csv',index = False)
```

```python
df_count_mean = df.groupby(['name']).count().reset_index()
df_count_mean.to_csv('df_count_mean.csv',index = False)
```
